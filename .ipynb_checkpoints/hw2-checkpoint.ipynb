- Visualize few images to check the data and labels
- Normalize (scale) the data
- Q: What is the number of features in this dataset:   64

```python
from sklearn import datasets
from sklearn import svm
from sklearn import metrics
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import pandas as pd
```

```python
digits = datasets.load_digits()
```

```python
digits
```

```python
imgTmp = digits.images[167,:,:].squeeze()
plt.imshow(imgTmp)
```

```python
scaler = StandardScaler()
scaler.fit(digits.data)
X = scaler.transform(digits.data)
```

### S2:
- Apply PCA with 30 components
- Calculate eigenvectors (PCA components) and eigenvalues for the data
- Plot variance explained by principal components
- Q: What is the variance explained by the first PCA component? 0.12033916097734805
- Q: What is the total variance explained by the first 5 PCA components? 0.4139794817637179

```python
NUM_COMP = 30

pca = PCA(n_components = NUM_COMP)
pcaMdl = pca.fit(X)
pcaCoor = pcaMdl.transform(X)
dfPCA = pd.DataFrame(data = pcaCoor, columns = ['PCA_' + s for s in np.arange(0, NUM_COMP).astype(str)])
```

```python
dfPCA.head()
```

```python
pcaVector = pcaMdl.components_
```

```python
dfPCA_vector = pd.DataFrame(data = pcaVector, columns = ['feature_' + s for s in np.arange(0, 64).astype(str)])
```

```python
IMG_NO = 377

f1 = digits.data[IMG_NO,:].reshape(imgSize)
plt.imshow(f1)
```

```python
plt.plot(pcaCoor[IMG_NO,0:NUM_COMP])
pcaCoor[IMG_NO,0:NUM_COMP]
```

```python
from numpy import linalg as LA

imgActual = digits.data[IMG_NO, :]

pic_377_1to30_list = []
for i in np.arange(1,31):
        
        imgRecon = np.dot(pcaCoor[IMG_NO, 0:i], pcaMdl.components_[0:i,:])
        imgRecon = scaler.inverse_transform(imgRecon)
        
        imgDiff = np.abs(imgActual - imgRecon)
        
        print('Num PCA comp: ' + str(i) + ', reconstruction error: ' + str(LA.norm(imgDiff.flatten())))
        pic_377_1to30_list.append(LA.norm(imgDiff.flatten()))
```

```python
IMG_NO = 105

imgRecon = np.dot(pcaCoor[IMG_NO,0:5], pcaMdl.components_[0:5,:])
imgRecon = scaler.inverse_transform(imgRecon)
f1 = imgRecon.reshape(imgSize)
plt.imshow(f1)
```

```python
imgActual = digits.data[IMG_NO, :]
imgRecon = np.dot(pcaCoor[105, 0:5], pcaMdl.components_[0:5,:])
imgRecon = scaler.inverse_transform(imgRecon)
imgDiff = np.abs(imgActual - imgRecon)
print(LA.norm(imgDiff.flatten()))
```

```python
compant_5_list = []
ind = 5
for i in range(0,len(digits.data)):
    imgActual = digits.data[i, :]
    imgRecon = np.dot(pcaCoor[i, 0:ind], pcaMdl.components_[0:ind,:])
    imgRecon = scaler.inverse_transform(imgRecon)
    imgDiff = np.abs(imgActual - imgRecon)
    compant_5_list.append(LA.norm(imgDiff.flatten()))
```

```python
digits_target_list = np.zeros([10])
digits_target_count = np.zeros([10])
for i in range(0,len(compant_5_list)):
    digits_target_count[digits.target[i]] = digits_target_count[digits.target[i]]+1
    digits_target_list[digits.target[i]] = digits_target_list[digits.target[i]]+compant_5_list[i]
```

```python
digits_ave_rec = []
for i in range(0,10):
#     print(digits_target_count[i]," ",digits_target_list[i])
    digits_ave_rec.append(digits_target_list[i]/digits_target_count[i])
digits_ave_rec
```

```python
print(digits_ave_rec.index(np.max(digits_ave_rec)))
print(digits_ave_rec.index(np.min(digits_ave_rec)))
print(compant_5_list.index(np.max(compant_5_list)))
print(compant_5_list.index(np.min(compant_5_list)))
```

### S4:

- Visualize reconstruction of first digit 3 image [1,3,5,7,9,11,13,15] PCA components

### S6:
- Randomly split the complete dataset into 2 (60% training 40% testing)
- Train an SVM classifier (rbf kernel, default parameters) using the training data
- Calculate testing prediction accuracy of the trained model
- Repeat the previous steps after applying PCA on the initial dataset and reducing the data dimensionality. Calculate prediction accuracy for 2, 5, 10 and 20 components
- Q: Which model obtains the highest classification accuracy?  20

```python
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier

X = digits.data
y = digits.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.4, random_state=42)
```

```python
mdlSVM = svm.SVC(kernel='rbf', probability=True, random_state=42,gamma=0.001, C=100.)
y_score = mdlSVM.fit(X_train, y_train).score(X_test, y_test)
y_score
```

```python
X = scaler.transform(X)
NUM_COMP = 30
pca = PCA(n_components = NUM_COMP)
pcaMdl = pca.fit(X)
pcaCoor = pcaMdl.transform(X)
```

```python
X_2 = []
X_5 = []
X_10 = []
X_20 = []
pcaList = [2,5,10,20]

for i in range(0,len(pcaCoor)):
    for j in range(0,4):
        imgRecon = np.dot(pcaCoor[i,0:pcaList[j]], pcaMdl.components_[0:pcaList[j],:])
        imgRecon = scaler.inverse_transform(imgRecon)
        if(j == 0):
            X_2.append(imgRecon)
        elif(j == 1):
            X_5.append(imgRecon)
        elif(j == 2):
            X_10.append(imgRecon)
        else:
            X_20.append(imgRecon)
```

```python
X_sum_set = [X_2,X_5,X_10,X_20]
```

```python
for i in range(0,4):
    X_train, X_test, y_train, y_test = train_test_split(X_sum_set[i], y, test_size=.4, random_state=42)
    y_score = mdlSVM.fit(X_train, y_train).score(X_test, y_test)
    print(pcaList[i]," score:", y_score)
```
